```python
# -*- coding: utf-8 -*-
"""
Created on Thu Nov 30 07:09:59 2023

@author: erimo
"""

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le dataset (assurez-vous que le fichier CSV est dans le mme rpertoire que ce script)
df = pd.read_csv('CS_Dataset.csv')

print(df)
# Afficher les premires lignes du dataset
print(df.head())

# Rsum statistique des variables numriques
print(df.describe())

# Vrifier les valeurs manquantes
print(df.isnull().sum())

# Explorer la distribution de l'ge
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution de l\'ge')
plt.show()

# Explorer la rpartition des genres
sns.countplot(x='Gender', data=df)
plt.title('Rpartition des genres')
plt.show()

# Explorer la relation entre le revenu et l'ducation
sns.scatterplot(x='Income', y='Education', data=df)
plt.title('Revenu par niveau d\'ducation')
plt.show()

# Explorer la corrlation entre les variables numriques
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matrice de corrlation')
plt.show()

# Explorer la relation entre le score de crdit et d'autres variables
sns.pairplot(df, hue='Credit Score')
plt.title('Relation entre le score de crdit et d\'autres variables')
plt.show()
```


La Diversit Dmographique: Le dataset reprsente un ventail diversifi de profils dmographiques.

ge Moyen: L'ge moyen des individus est d'environ 38 ans, avec un minimum de 25 ans et un maximum de 53 ans.

Revenu: Les revenus varient de 25 000 $  162 500 $, avec un revenu moyen de 83 765 $.

Nombre d'Enfants: En moyenne, les individus ont moins d'un enfant, avec un maximum de trois.

Variables Supplmentaires: Le jeu de donnes inclut des informations sur l'ducation, l'tat matrimonial, la proprit du logement et les scores de crdit.

Le march prsente une forte correlation entre les revenus et l'ge ce qui sous-entend que les personnes les plus ages devraient avoir un revenu plus grand.

Cette tude de march permet galement de nous dire que les personnes ayant un grand score de crdit sont les personnes ages de plus de 38 ans et ou qui ont au plus 1 enfants.

On se rend compte cependant rapidement que les personnes ayant plus de 38 ans n'ont pas forcment plusieurs enfants. Ce qui est important  savoir lors du lancement d'un ventuel produit.
